# COUNT FILMS

imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

get dataframes from excel

```python
xl = pd.read_excel('blank-check-data.xlsx', sheet_name=None, index_col='CODE')
df_miniseries = xl['MINISERIES']
df_films = xl['FILMS']
```

get count

```python
df_miniseries['FILMCOUNT'] = [len(df_films[df_films['MINISERIES'] == code]) for code in df_miniseries.index.values]
```

print sorted

```python
print(df_miniseries.sort_values(['FILMCOUNT'], ascending=False)[['NAME', 'FILMCOUNT']])
```

```python
df_miniseries.plot(kind='bar', y='FILMCOUNT')
plt.show()
```

```python
df_miniseries.plot(kind='hist', y='FILMCOUNT')
plt.show()
```
