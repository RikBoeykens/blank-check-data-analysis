```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xl = pd.read_excel('blank-check-data.xlsx', sheet_name=None, index_col='CODE')
df_miniseries = xl['MINISERIES']
df_films = xl['FILMS']
```

```python
df_director_miniseries = df_miniseries[df_miniseries['DIRECTOR']]
```

```python
df_director_films = df_films[np.logical_and(df_films['MINISERIES'].isin(df_director_miniseries.index.values), df_films['NONDIRECTOR']==False)]
```

```python
df_director_films.boxplot(by='MINISERIES', column=['RUNTIME'])
plt.style.use("classic")
plt.suptitle('')
plt.ylabel('')
plt.xlabel('')
plt.locator_params(axis='y', nbins=24)
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = [16,9]
plt.show()
```
