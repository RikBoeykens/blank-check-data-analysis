```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xl = pd.read_excel('blank-check-data.xlsx', sheet_name=None, index_col='CODE')
df_miniseries = xl['MINISERIES']
df_films = xl['FILMS']
```

```python
df_director_miniseries = df_miniseries[df_miniseries['DIRECTOR']]
```

```python
plt.style.use("dark_background")
for index, row in df_director_miniseries.iterrows():
    df_miniseries_films = df_films[np.logical_and(np.logical_and(np.logical_and(df_films['MINISERIES']==index, df_films['BUDGET']!=0), df_films['REVENUE']!=0), df_films['NONDIRECTOR']==False)].sort_values("RELEASEDATE")
    if(len(df_miniseries_films)) :
        names = list(df_miniseries_films['NAME'])
        budgets = list(df_miniseries_films['BUDGET'])
        budgetsTimesTwoAndHalf = list(df_miniseries_films['BUDGET']* 5 /2) 
        revenues = list(df_miniseries_films['REVENUE'])
        
        df = pd.DataFrame({'budget': budgets, 'budget * 2.5': budgetsTimesTwoAndHalf, 'revenue': revenues}, index=names)
        ax = df.plot.bar(title=index)
        ax.ticklabel_format(style='plain', axis='y')
        plt.show()
```
